```python
df=df.drop_duplicates(subset='Order_ID',keep='first')
df
```

```python
num_cols=['Customer_Age','Price','Quantity','Delivery_Time_mins','Rating']
df
```

```python
for col in num_cols:
  df[col]=df[col].fillna(df[col].median())
```

```python
cat_cols=['City','Payment_Method','Order_Status']
df
```

```python
for col in cat_cols:
  df[col]=df[col].fillna(df[col].mode()[0])
```

```python
df['Category'].unique()
```

```python
df['Category']=df['Category'].str.strip()
```

```python
df['Category']=df['Category'].str.title()
```

```python
df['Category']=df['Category'].str.strip().str.title()
```

```python
df['Delivery_Time_mins'].describe()
```

```python
df=df[(df['Delivery_Time_mins']>=0)&(df['Delivery_Time_mins']<=60)]
df
```

```python
df=df[(df['Rating']>=1)&(df['Rating']<=5)]
df
```

```python
df.dtypes
```

```python
df['Customer_Age']=df['Customer_Age'].astype('int64')
df
```

```python
df['Quantity']=df['Quantity'].astype('int64')
df
```

```python
df['Rating']=df['Rating'].astype('int64')
df
```

```python
df['Total_Amount']=df['Price']*df['Quantity']
df
```
